```python
trning["city"].fillna(value = "UP", inplace = True)
trning['winner'].fillna(value = 'Draw', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
trning["city"]=encoder.fit_transform(trning["city"])
trning["team1"]=encoder.fit_transform(trning["team1"])
trning["team2"]=encoder.fit_transform(trning["team2"])
trning["winner"]=encoder.fit_transform(trning["winner"].astype(str))
trning["toss_winner"]=encoder.fit_transform(trning["toss_winner"])
trning["toss_decision"]=encoder.fit_transform(trning["toss_decision"])
trning["venue"]=encoder.fit_transform(trning["venue"])
```

```python
trning
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
ids = trning[["id"]]
trning.drop(columns = "id",inplace = True)
```

```python
X = trning[["city","team1","team2","toss_winner","toss_decision","venue"]]
Y = trning[["winner"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lf = LogisticRegression()
lf.fit(x_train, y_train)
pred = lf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred,y_test)
acc
```

```python
rf = RandomForestClassifier(n_estimators=300)
rf.fit(x_train, y_train)
pred1 = rf.predict(x_test)
acc1 = accuracy_score(pred1,y_test)
acc1
```
