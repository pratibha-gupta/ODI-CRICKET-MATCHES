```python
df1.drop(['Scorecard'],inplace = True,axis = 1)
df1.drop(["Match Date"], inplace = True,axis = 1)
```

```python
s = pd.concat([df1['Team 1'],pd.Series(['U.S.A.','no result','ICC World XI','tied'])])
#s1 = pd.concat([s,['no result']])
```

```python
s.drop_duplicates(inplace = True)
```

```python
#x = pd.get_dummies(df1)
#y = pd.get_dummies(y)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(s)
#le.fit(df1['Team 2'])
```

```python
df1['Team 1'] = le.transform(df1['Team 1'])
df1['Team 2'] = le.transform(df1['Team 2'])
df1['Winner'] = le.transform(df1['Winner'])
```

```python
df1
```

```python
y = df1[['Winner']]
```

```python
df1.drop(['Winner','Margin','Ground'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test = train_test_split(df1, y, test_size=0.20,random_state=100)
```

```python
rf = RandomForestClassifier() 
rf.fit(x_train, y_train)
```

```python
from sklearn import metrics
train_pred = rf.predict(x_train)
test_pred = rf.predict(x_test)
```

```python
train_pred
```

```python
from sklearn.metrics import accuracy_score
acc_score_train = accuracy_score(y_train,train_pred)
acc_score_test = accuracy_score(y_test,test_pred)
```

```python
print(acc_score_train)
print(acc_score_test)
```
